```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def getStrategy(regretSum, strategySum):  #input regretSum, output strategy and update strategySum
    strategy = np.array([0.0, 0.0, 0.0])
    normalizingSum = 0.0;
    for i_action in range(3):
        strategy[i_action] = regretSum[i_action] if regretSum[i_action] > 0 else 0
        normalizingSum += strategy[i_action]
    for i_action in range(3):
        if normalizingSum > 0:
            strategy[i_action] /= normalizingSum
        else:
            strategy[i_action] = 1.0 / 3;
        strategySum[i_action] += strategy[i_action]
    return strategy, strategySum;

#input strategy and output avgstrategy, normalized to sum to 1
def getAvgStrategy(strategySum):
    avgStrategy = np.array([0.0, 0.0, 0.0])
    normalizingSum = 0.0;
    for i_action in range(3):
        normalizingSum += strategySum[i_action]
    for i_action in range(3):
        if normalizingSum > 0:
            avgStrategy[i_action] = strategySum[i_action] / normalizingSum
        else:
            avgStrategy[i_action] = 1.0 / 3;
    return avgStrategy

def getAction(strategy):
    rr = random.random()
    cumulativeProbability = 0
    return np.searchsorted(np.cumsum(strategy), rr)


def train(iterations):
    stratPlotFreq = 10;
    convPlotFreq = 10;
    
    oppStrategy = np.array([0.34, 0.33, 0.33])
    historyAvgStrategyRock=[]
    historyAvgStrategyPaper=[]
    historyAvgStrategyScissor=[]
    historyExploitiveness=[]
    historyRegretRock=[]
    historyRegretPaper=[]
    historyRegretScissor=[]
    
    ## the strategy that converges, updated when calling getStrategy() by strategy
    strategySum = np.array([0.0, 0.0, 0.0])  
    ##regret of each decision, updated each iteration
    regretSum = np.array([0.0, 0.0, 0.0])
    
    actionUtility = np.array([0.0, 0.0, 0.0])
    for i_iteration in range(iterations):
        #debug print
        #print( "average strategy:", getAvgStrategy(strategySum))
        #print( "strategySum:" , strategySum)
        #print( "regretSum:", regretSum)
        strategy, strategySum = getStrategy(regretSum, strategySum)
        myAction = getAction(strategy)
        otherAction = getAction(oppStrategy)
        #print( "my action:", myAction, " other action:", otherAction )
        actionUtility = getActionUtility(myAction, otherAction)
        #print( "actionUtility is: ", actionUtility)
        regretSum = accumulateActionRegret(actionUtility, myAction, regretSum)
        
        #log evolution of average strategy and regret
        if (i_iteration % stratPlotFreq == 0):
            historyAvgStrategyRock.append(getAvgStrategy(strategySum)[0])
            historyAvgStrategyPaper.append(getAvgStrategy(strategySum)[1])
            historyAvgStrategyScissor.append(getAvgStrategy(strategySum)[2])
            historyRegretRock.append(regretSum[0])
            historyRegretPaper.append(regretSum[1])
            historyRegretScissor.append(regretSum[2])
            
        if (i_iteration % convPlotFreq == 0):
            historyExploitiveness.append(getExploitativeValue(getAvgStrategy(strategySum),oppStrategy))
        
    print( "average strategy:", getAvgStrategy(strategySum))
    #print( "average strategy fold:", historyAvgStrategyRock, " type:", type(historyAvgStrategyRock), " length:", len(historyAvgStrategyRock))
    #print( "average strategy call:", historyAvgStrategyPaper, " type:", type(historyAvgStrategyPaper), " length:", len(historyAvgStrategyPaper))
    #print( "average strategy raise:", historyAvgStrategyScissor, " type:", type(historyAvgStrategyScissor), " length:", len(historyAvgStrategyScissor))
    
    plt.stackplot(np.arange(0,iterations,stratPlotFreq).tolist(), \
                  historyAvgStrategyRock, historyAvgStrategyPaper, historyAvgStrategyScissor, \
                  labels=['rock','paper','scissor'])
    plt.legend(loc='lower right')
    plt.xlabel('iterations')
    plt.ylabel('strategy')
    plt.show()

    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyRegretRock, label="Rock")
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyRegretPaper, label="Paper")
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyRegretScissor, label="Scissor")

    #plt.yscale('log')
    plt.xlabel('iterations')
    plt.ylabel('regret')
    plt.legend(loc='lower left')
    plt.show()
    
    plt.plot(np.arange(0,iterations,convPlotFreq).tolist(),historyExploitiveness)
    plt.xlabel('iterations')
    plt.ylabel('ev')

    plt.show()
    return getAvgStrategy(strategySum)
    
def train2p(iterations):
    stratPlotFreq = 10;
    convPlotFreq = 10;
    
    historyAvgStrategyRock=[]
    historyAvgStrategyPaper=[]
    historyAvgStrategyScissor=[]
    historyExploitiveness=[]
    historyRegretRock=[]
    historyRegretPaper=[]
    historyRegretScissor=[]
    historyOppAvgStrategyRock=[]
    historyOppAvgStrategyPaper=[]
    historyOppAvgStrategyScissor=[]
    historyOppRegretRock=[]
    historyOppRegretPaper=[]
    historyOppRegretScissor=[]
    
    ## the strategy that converges, updated when calling getStrategy() by strategy
    strategySum = np.array([0.0, 0.0, 0.0])  
    ##regret of each decision, updated each iteration
    regretSum = np.array([0.0, 0.0, 0.0])
    oppStrategySum = np.array([0.0, 0.0, 0.0])  
    oppRegretSum = np.array([0.0, 0.0, 0.0])

    actionUtility = np.array([0.0, 0.0, 0.0])
    oppActionUtility = np.array([0.0, 0.0, 0.0])
    
    for i_iteration in range(iterations):
        #debug print
        #print( "average strategy:", getAvgStrategy(strategySum))
        #print( "strategySum:" , strategySum)
        #print( "regretSum:", regretSum)
        strategy, strategySum = getStrategy(regretSum, strategySum)
        myAction = getAction(strategy)
        
        oppStrategy, oppStrategySum = getStrategy(oppRegretSum, oppStrategySum)
        oppAction = getAction(oppStrategy)
        
        #debug print
        #print( "my action:", myAction, " opp action:", oppAction )
        
        actionUtility = getActionUtility(myAction, oppAction)
        OppActionUtility =  getActionUtility(oppAction, myAction)
        
        #debug print
        #print( "actionUtility is: ", actionUtility)
        #print("oppo action utility: ", OppActionUtility)
        
        #update regret
        regretSum = accumulateActionRegret(actionUtility, myAction, regretSum)
        oppRegretSum = accumulateActionRegret(OppActionUtility, oppAction, oppRegretSum)
        
        if (i_iteration % stratPlotFreq == 0):
            historyAvgStrategyRock.append(getAvgStrategy(strategySum)[0])
            historyAvgStrategyPaper.append(getAvgStrategy(strategySum)[1])
            historyAvgStrategyScissor.append(getAvgStrategy(strategySum)[2])    
            historyRegretRock.append(regretSum[0])
            historyRegretPaper.append(regretSum[1])
            historyRegretScissor.append(regretSum[2])
            historyOppAvgStrategyRock.append(getAvgStrategy(oppStrategySum)[0])
            historyOppAvgStrategyPaper.append(getAvgStrategy(oppStrategySum)[1])
            historyOppAvgStrategyScissor.append(getAvgStrategy(oppStrategySum)[2])
            historyOppRegretRock.append(oppRegretSum[0])
            historyOppRegretPaper.append(oppRegretSum[1])
            historyOppRegretScissor.append(oppRegretSum[2])
            
        if (i_iteration % convPlotFreq == 0):
            historyExploitiveness.append(getExploitativeValue(getAvgStrategy(strategySum),getAvgStrategy(oppStrategySum)))
        
    print( "average strategy:", getAvgStrategy(strategySum))
    #print( "average strategy fold:", historyAvgStrategyRock, " type:", type(historyAvgStrategyRock), " length:", len(historyAvgStrategyRock))
    #print( "average strategy call:", historyAvgStrategyPaper, " type:", type(historyAvgStrategyPaper), " length:", len(historyAvgStrategyPaper))
    #print( "average strategy raise:", historyAvgStrategyScissor, " type:", type(historyAvgStrategyScissor), " length:", len(historyAvgStrategyScissor))

    
    #print player 1 strategy
    plt.stackplot(np.arange(0,iterations,stratPlotFreq).tolist(), \
                  historyAvgStrategyRock, historyAvgStrategyPaper, historyAvgStrategyScissor, \
                  labels=['rock','paper','scissor'])
    plt.legend(loc='lower right')
    plt.xlabel('iterations')
    plt.ylabel('strategy')
    plt.title('player 1 strategy')
    plt.show()
    
    
    #print player 2 strategy
    plt.stackplot(np.arange(0,iterations,stratPlotFreq).tolist(), \
                  historyOppAvgStrategyRock, historyOppAvgStrategyPaper, historyOppAvgStrategyScissor, \
                  labels=['rock','paper','scissor'])
    plt.legend(loc='lower right')
    plt.xlabel('iterations')
    plt.ylabel('strategy')
    plt.title('player 2 strategy')
    plt.show()
    
    #player 1 regret history
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyRegretRock, label="Rock")
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyRegretPaper, label="Paper")
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyRegretScissor, label="Scissor")
    #plt.yscale('log')
    plt.xlabel('iterations')
    plt.ylabel('regret')
    plt.legend(loc='lower left')
    plt.show()
    
    
    #player 2 regret history
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyOppRegretRock, label="Rock")
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyOppRegretPaper, label="Paper")
    plt.plot(np.arange(0,iterations,stratPlotFreq).tolist(), historyOppRegretScissor, label="Scissor")
    plt.xlabel('iterations')
    plt.ylabel('regret')
    plt.legend(loc='lower left')
    plt.show()    


    
    #print EV of player1
    plt.plot(np.arange(0,iterations,convPlotFreq).tolist(),historyExploitiveness)
    plt.xlabel('iterations')
    plt.ylabel('ev')
    plt.show()


    
    
    
    return getAvgStrategy(strategySum)



#action utility returns the value of each action at current node

def getActionUtility(myAction, otherAction):
    actionUtility=np.array([0.0,0.0,0.0])
    actionUtility[0] = getOutcome(0, otherAction)
    actionUtility[1] = getOutcome(1, otherAction)
    actionUtility[2] = getOutcome(2, otherAction)
    return actionUtility


# 0 = rock, 1 = paper, 2 = scissors
def getOutcome(myAction, oppAction):
    if (myAction == oppAction):
        return 0
    elif ((myAction == 0 and oppAction == 1) or (myAction == 1 and oppAction == 2)):
        return -1
    elif ((myAction == 2 and oppAction == 1) or (myAction == 1 and oppAction == 0)):
        return 1
    elif ((myAction == 0 and oppAction == 2)):
        return 100
    elif ((myAction == 2 and oppAction == 0)):
        return -100
    
    return 0

#accumulate regret with respect to my action
def accumulateActionRegret(actionUtility, myAction, regretSum):
    for i_action in range(3):
        regretSum[i_action] += actionUtility[i_action] - actionUtility[myAction]
    return regretSum

#test strategy ev
def getExploitativeValue(myStrategy, oppStrategy):
    outcomeSum = 0.0;
    
    #normalized game table
    outcomeMatrix = np.zeros(9).reshape((3, 3))
    for i in range(3):
        for j in range(3):
            outcomeMatrix[i][j]= getOutcome(i,j)
    
    #debug 
    #print("outcome matrix:", outcomeMatrix)
    #frequency occurance
    frequencyMatrix = np.outer(myStrategy, oppStrategy)
    
    
    return sum(sum(np.multiply(outcomeMatrix, frequencyMatrix)))
```
